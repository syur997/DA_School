```python
import matplotlib.pyplot as plt

font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=15, fontdict=font1)
plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)
plt.show()
```

## 축 범위 지정

```python
# 축 범위 얻기
# xlim(), ylim() 함수는 그래프 영역에 표시되는 x축, y축의 범위를 각각 반환한다.
# axis() 함수는 그래프 영역에 표시되는 x, y축의 범위를 반환한다.

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis('scaled')
print(axis_range)

plt.show()
```

## 선 종류 지정하기

```python
# 선 종류 지정하기

import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')    # '-' 형태로 뽑겠다.
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

```python
# 위 그래프와 똑같지만 linestyle로도 구현 가능

import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
```

```python
# linestyle을 정하면 튜플에 따라서 커스터마이즈 할 수 있다.

# 튜플을 사용해서 선의 종류를 커스터마이즈할 수 있습니다.

# 예를 들어, (0, (1, 1))은 ‘dotted’와 같고, (0, (5, 5))는 ‘dashed’와 같습니다.

# 또한 (0, (3, 5, 1, 5))는 ‘dashdotted’와 같습니다.

# 숫자를 바꿔가면서 다양한 선의 종류를 만들어보세요.

# 결과는 아래와 같습니다.
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], linestyle=(0, (1, 1)), color='C0', label='(0, (1, 1))')
plt.plot([1, 2, 3], [3, 3, 3], linestyle=(0, (1, 5)), color='C0', label='(0, (1, 5))')
plt.plot([1, 2, 3], [2, 2, 2], linestyle=(0, (5, 1)), color='C0', label='(0, (5, 1))')
plt.plot([1, 2, 3], [1, 1, 1], linestyle=(0, (3, 5, 1, 5)), color='C0', label='(0, (3, 5, 1, 5))')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

## 선 끝 모양 지정하기



```python
# 선 끝 모양 지정하기 ( solid_capstyle ), linewidth로 굵기 조절 가능
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', linewidth=10,
      solid_capstyle='butt', color='C0', label='solid+butt')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='solid', linewidth=5,
      solid_capstyle='round', color='C0', label='solid+round')

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=5,
      dash_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashed', linewidth=10,
      dash_capstyle='round', color='C1', label='dashed+round')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

## 그래프의 영역 나타내기

```python
# 그래프의 영역을 나타내기 ( fill_between )
# fill_between() 함수에 x[1:3], y[1:3]를 순서대로 입력하면,
# 네 점 (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역이 채워집니다.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)

plt.show()
```

# 막대그래프 그리기

```python
# 1. 막대그래프 그리기 (plt.bar)

import matplotlib.pyplot as plt
import numpy as np

x = [0,1,2]
years = ['2018', '2019', '2020']
values = [100, 400, 900]

# plt.bar
plt.bar(x, values) 
plt.xticks(x, years)

plt.show()
```

```python
# 이 예제는 연도별로 변화하는 값을 갖는 데이터를 막대 그래프로 나타냅니다.

# [0,1,2] 주어진 범위와 간격에 따라 균일한 값을 갖는 리스트를 만들었습니다.

# years는 X축에 표시될 연도이고, values는 막대 그래프의 y 값 입니다.

# 먼저 plt.bar() 함수에 x 값 [0, 1, 2]와 y 값 [100, 400, 900]를 입력해주고,

# xticks()에 x와 years를 입력해주면, X축의 눈금 레이블에 ‘2018’, ‘2019’, ‘2020’이 순서대로 표시됩니다.

# 결과는 아래와 같습니다.
```

## 막대그래프의 색깔 지정

```python
# 막대그래프의 색깔 지정 (plt.bar - color)

import matplotlib.pyplot as plt
import numpy as np

x = [0,1,2]
years = ['2018', '2019', '2020']
values = [100, 400, 900]

# plt.bar에서 color 지정
plt.bar(x, values, color='pink') 
plt.xticks(x, years)

plt.show()
```

```python
# 막대별로 색상 다르게 표현

import matplotlib.pyplot as plt
import numpy as np

x = [0,1,2]
years = ['name1', 'name2', 'name3']
values = [100, 400, 900]
colors = ['C2', 'pink', 'dodgerblue']

plt.bar(x, values, color=colors) 
plt.xticks(x, years)

plt.show()
```

## 엣지 표시하기

# 세 번째 데이터 실습

```python
# 남녀성비 그래프
import csv
f = open('/content/dataset_example3_new.csv')
data = csv.reader(f)

female = [] # 여성 인구정보를 담을 리스트
male = [] # 남성 인구정보를 담을 리스트

for row in data :
  if '제주특별자치도' in row[0] :
    for i in row[16:27] :  # 1에서 100까지
      i = i.replace(',', '')
      male.append(int(i))  # 남성데이터는 15~27
    for i in row[29:] :
      i = i.replace(',' ,'')
      female.append(int(i))  # 여성데이터는 30~번까지


import matplotlib.pyplot as plt

plt.bar(range(11), male)
plt.bar(range(11), female)
plt.legend()
plt.show()
f.close()
```

```python
# 위 그래프를 barh 그래프로 변경해보기

import csv
f = open('/content/dataset_example3_new.csv')
data = csv.reader(f)

female = [] # 여성 인구정보를 담을 리스트
male = [] # 남성 인구정보를 담을 리스트

for row in data :
  if '제주특별자치도' in row[0] :
    for i in row[16:27] :  # 1에서 100까지
      i = i.replace(',', '')
      male.append(-int(i))  # 남성데이터는 15~27
    for i in row[29:] :
      i = i.replace(',' ,'')
      female.append(int(i))  # 여성데이터는 30~번까지


import matplotlib.pyplot as plt

plt.barh(range(11), male)
plt.barh(range(11), female)
plt.legend()
plt.show()
f.close()
```

```python
# 문제
# 리스트 3개(연도 / 서울 / 수도권)를 만들어서 집어넣기
import csv

with open('example_4_all.csv') as f :
  reader = csv.reader(f)
  연도 = []
  서울 = []
  수도권 = []

  for i in reader :
    if i[0] != "구분":
      연도.append(i[0])
      서울.append(float(i[1]))
      수도권.append(float(i[2]))

print(연도)
print(서울)
print(수도권)
```

```python
# 그래프 문제

# 연도가 x축의 이름
# 서울과 수도권을 2개의 그래프로 표현하기

import matplotlib.pyplot as plt

plt.plot(range(14), 서울, label = '서울')
plt.plot(range(14), 수도권, label = '수도권')

plt.xticks(range(14), 연도, rotation = 45)

plt.show()
```

```python
import matplotlib.pyplot as plt
# x축에 년도, y축에 인구수가 있는 선 그래프 만들기
plt.plot(range(14), 서울, color='green', marker='o', linestyle='solid', label='s')
plt.plot(range(14), 수도권, color='red', marker='o', linestyle='solid', label='g')

plt.xticks(range(14), 연도, rotation=45)

plt.title("Population Change")
plt.ylabel("Percentage")

plt.legend()
plt.show()
```
