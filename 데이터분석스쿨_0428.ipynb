## 함수 안의 함수

```python
# 함수 안의 변수나 함수는 함수가 끝나면 휘발된다.
# 메모리에서 사라짐
# 함수 안 함수(f1)는 밖에서 호출되지 않음

def f2(x, y):
  def f1(x):
    print('world1')
    return x + x
  print('hello1')
  z = f1(10)
  return x + y + z

# a = f1(1)  -> 함수 안 함수는 밖에서 호출 안됨
b = f2(3, 4)
print(b)

# print(z) -> 함수가 끝났기 때문에 x, y, z에 접근이 안됨
```

## 재귀함수

* 내가 나를 반복적으로 호출하는 함수
* 대부분의 코드는 for문으로 대체할 수 있다.
* 재귀함수는 효율이 매우 좋지 않다.
* 다만, 메모라이징 기법을 사용하면 좋다.
* 다이나믹 프로그래밍 등 필수적으로 사용되는 곳이 있기도 하다.

```python
# #sys.setrecursionlimit -> 제한을 늘릴 수 있음
# def 숫자출력():
#     print(1)
#     return 숫자출력()
# 숫자출력()

# 이거 실행하면 1이 계속 출력됨 -> 재귀함수
# 무한 반복에 종료조건을 걸 수는 있음
```

```python
# 회문(앞뒤가 같은 문자, 토마토, 오디오, 역삼역, 기러기 등)
# factorial : 5 -> 5 * 4 * 3 * 2 * 1

def factorial(n):
  if n == 1 :
    return n
  return n * factorial(n-1)

factorial(5)

#                     return 값
# factorial(5) == 5 * factorial(4) == 5 * 24
# factorial(4) == 4 * factorial(3) == 4 * 6
# factorial(3) == 3 * factorial(2) == 3 * 2
# factorial(2) == 2 * factorial(1) == 2 * 1
# factorial(1) == 1
```

```python
# 위의 재귀함수문을 아래의 for문 range로 간단하게 구현 가능

result = 1
for i in range(1, 6):
    result *= i

result
```

```python
s = 'leehojun'
result = ''
for i in s:
    result = i + result  # i가 앞에 있으니까 거꾸로 출력됨 / 올바르게 출력하려면 result + i

result
```

```python
def string_reverse(s):
  if len(s) == 1 :   # 얼리리턴(코드가 매우 길면 앞쪽에서 return 해주는게 좋음)
    return s
  return string_reverse(s[1:]) + s[0]

string_reverse('hello')

# string_reverse('hello') == string_reverse('ello') + 'h' == 'olle' + 'h' == 'olleh'
# string_reverse('ello') == string_reverse('llo') + 'e' == 'oll' + 'e' == 'olle'
# string_reverse('llo') == string_reverse('lo') + 'l' == 'ol' + 'l' == 'oll'
# string_reverse('lo') == string_reverse('o') + 'l' == 'o' + 'l' == 'ol'
# string_reverse('o') == 'o'
```

```python
# 재귀함수가 얼마나 비효율적인지 보여주는 코드

def fib(num):
    if num == 0:
        return 0
    if num == 1 or num == 2:
        return 1
    else:
        return fib(num-1) + fib(num-2)

fib(35)

# fib(35)를 했을 때, 3초가 나옴
```

```python
# 위와 같은 코드지만 데코레이터..?를 사용했을 때 코드

from functools import lru_cache

@lru_cache(maxsize=None)
def fib(num):
    if num == 0:
        return 0
    if num == 1 or num == 2:
        return 1
    else:
        return fib(num-1) + fib(num-2)

fib(40)
```

```python
# 함수를 이용하여 계산기 프로그램을 만들기

def plus(num1, num2):
  return num1 + num2

def minus(num1, num2):
	return num1 - num2

def multiply(num1, num2):
	return num1 * num2

def divide(num1, num2):
	return num1 / num2

print('더하기 : ',plus(10,5))
print('빼기 : ',minus(10,5))
print('곱하기 : ',multiply(10,5))
print('나누기 : ',divide(10,5))
```

```python
# 강사님 버전(try, except)

def plus(num1, num2):
    try:
        result = int(num1) + int(num2)
        return result
    except:
        return float("inf")  # float("inf"):양의 무한대수 / return "반환할 수 없습니다" 도 가능

print(f'더하기 : {plus(10, 5)}')
print(f'더하기 : {plus("10", "5")}')
print(f'더하기 : {plus("10", 5)}')
print(f'더하기 : {plus(10, "5")}')
print(f'더하기 : {plus("abc", 5)}')
print(f'더하기 : {plus(10, "abc")}')
```

```python
# 강사님 버전 (lambda)

plus = lambda x, y: x + y
minus = lambda x, y: x - y
multiply = lambda x, y: x * y
divide = lambda x, y: x / y

print(f'더하기 : {plus(10, 5)}')
print(f'빼기 : {minus(10, 5)}')
print(f'곱하기 : {multiply(10, 5)}')
print(f'나누기 : {divide(10, 5)}')
```

```python
# 재귀함수를 이용하여 x의 n제곱이 되는 함수 만들기

def pow(x, n):
    x **= n
    return x

pow(2,3)

# 내가 푼건 그냥 함수인가봄;
```

```python
# 강사님이 쓰신 코드

def 제곱(x, n):
  if n == 0 :
    return 1
  return x * 제곱(x, n-1)

제곱(3, 4)
```
