```python
# ì •ê·œí‘œí˜„ì‹ê³¼ ì¼ì¹˜í•˜ëŠ” ë¶€ë¶„ í™•ì¸

# re ëª¨ë“ˆì˜ serachí•¨ìˆ˜ëŠ” ì •ê·œí‘œí˜„ì‹ê³¼ ì¼ì¹˜í•˜ëŠ” ì²« ë²ˆì§¸ ìœ„ì¹˜ ì •ë³´ê°€ ì €ì¥ëœ Match ê°ì²´ë¥¼ ë°˜í™˜
# Match ê°ì²´ë¥¼ ì‚¬ìš©í•˜ë©´ ì¼ì¹˜í•˜ëŠ” ìœ„ì¹˜ì˜ ë¬¸ìì—´ê³¼ í•¨ê»˜ ì‹œì‘ ë° ì¢…ë£Œ ìœ„ì¹˜ ì •ë³´ í™•ë³´ ê°€ëŠ¥

# matchê°ì²´
# start , end , group, groups íŠœí”Œ ê·¸ë£¹

import re

text = 'Error should never pass silently'

m_obj = re.search(r"p...", text)

print(m_obj.group()) # ì¼ì¹˜ ë¬¸ìì—´ ()ë¡œ ë¬¶ì¸ ê²ƒì¸ ì—†ìŒ
print(m_obj.start()) # ì‹œì‘ ì¸ë±ìŠ¤
print(m_obj.end()) # ë ì¸ë±ìŠ¤
```

```python
# 1ë²ˆ ê·¸ë£¹ - n....  n ë‹¤ìŒì— 4ê¸€ìê°€ ì•„ë¬´ê±°ë‚˜
# 2ë²ˆ ê·¸ë£¹ - p...  n ë‹¤ìŒì— 3ê¸€ìê°€ ì•„ë¬´ê±°ë‚˜

import re

text = 'Error should never pass silently'

m_obj = re.search(r"(n....) (p...)", text)
print(m_obj.group())
print(m_obj.groups())
```

```python
# Greedy, Lazy ì‚¬ìš©í•˜ê¸°
# ì •ê·œí‘œí˜„ì‹ì´ íŒ¨í„´ì— ì¼ì¹˜í•˜ëŠ” ë¬¸ìì—´ì„ ì¶”ì¶œí•  ë•Œ
# ì¼ì¹˜í•˜ëŠ” ìµœëŒ€í•œì˜ ë²”ìœ„ë¥¼ ì¶”ì¶œí•˜ëŠ” ìƒíƒœë¥¼ greedy(íƒìš•)ì´ë¼ê³  í•œë‹¤.

# ìµœì†Œí•œì€ Lazyë¼ê³  í•˜ëŠ”ë° íŒŒì´ì¬ì€ ê¸°ë³¸ì ìœ¼ë¡œ greedyí•˜ì§€ë§Œ
# ?ë¥¼ ë¶™ì´ëŠ” ìˆœê°„ Lazyí•´ì§„ë‹¤. 

import re

text = 'In the face of ambiguity, refuse the temptation to guess'

match_list = re.findall(r"t.*\s", text) # të¡œ ì‹œì‘í•˜ëŠ” ì•„ë¬´ ë¬¸ìë¥¼ ê³µë°± í¬í•¨ ì¶”ì¶œ
print(match_list)

# ê²°ê³¼ë¥¼ ë³´ë©´ to ë‹¤ìŒ ê³µë°±ê¹Œì§€ ì¶”ì¶œ
# ê³µë°±ê¹Œì§€ ê°€ì ¸ì˜¤ë¼ê³  í•´ì„œ ë§ˆì§€ë§‰ ê³µë°±ê¹Œì§€ ë‹¤ ê°€ì ¸ì˜¨ ê²°ê³¼ë¥¼ ë³´ì„
```
