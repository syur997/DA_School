# 조건문 복습

조건문은 다음과 같이 복습하려고 합니다.

1. 단일 조건문
2. If - else 조건문
3. If - elif - else 조건문
4. 다중 중첩 조건문
5. input과 함께 사용하는 조건문

[기억해야할것]

1. if 뒤에는 참과 거짓을 판단하는 조건이 와야함
2. 조건문이 끝나면 반드시 콜론으로 끝나야함
3. if 다음 문장은 들여쓰기를 사용해야함
4. if 조건에 부합하지 않으면 elif나 else를 사용한다. 보통 조건에 모두 부합하지 않을때 else를 사용한다.

[연산자]

* x < x가 작다
* x > x가 크다
* x == y 같다
* x != y 다르다
* x >= y 크거나 같다
* x <= y 작거나 같다

[논리연산자]
2개이상의 조건을 사용하고 싶을때 사용

* and
* or
* not

# 단일 조건문



```python
# 1.1 단일 조건문 예시

# if 조건 :
#   실행할 명령

# 예제1. 숫자열을 입력할 경우

total = 5

if total > 0:
  print(f'{total}명이 입장하였습니다.')
```

```python
# 예제2. 숫자열 입력시 오류

total = input('몇 명이 들어왔나요?')  # input으로 받으면 string

if total > 0 :  # input으로 받은 string과 0인 int를 비교하면 에러가 뜸
  print(f'{total} 인이 입장하였습니다.')
```

```python
# 예제3. 문자열을 입력할 때

wheather = input('오늘 날씨는 어떤가요?')

if wheather != '' :  # 공백일 때는 if문 조건이 성립되지 않아서 print문이 실행되지 않음
  print(f'오늘 날씨는 {wheather}입니다.')
```

```python
# 예제4. 2개 이상 조건일 때
# 1. 오늘 날씨가 어떤가요?
# 2. 오늘 기온이 어떤가요?

wheater = input('오늘 날씨가 어떤가요?')
temp = input('오늘 기온이 어떤가요?')

if wheater != '' :
  print(f'오늘 날씨는 {wheather}입니다.')
  if temp != '' :
    print(f'오늘 기온은 {temp}입니다.')
```

```python
# 위의 예제4번. 강사님ver.

wheater = input('오늘 날씨가 어떤가요?')
temp = int(input('오늘 기온이 어떤가요?'))  # if문에서 0과 비교하기 위해 int로 바꿔주기

if wheather != '' and temp > 0 :
  print(f'오늘 날씨는 {wheather}입니다. 오늘 기온은 {temp}입니다.')
```

```python
# 예제5. if문을 쓸 때 헷갈리는 것

job = input('직업이 있나요?')

if job == '예' :
  print('직업이 있습니다.')
```

# 다중 조건문 - if else문
* 조건이 2개인 경우에는 보통 if else를 사용하지만 if elif를 사용하기도 한다.

[주의]
* else 사용할 때는 누락되는 조건이 없는지 확인해봐야한다.

```python
# if 조건1 :
#   실행문장 1
# else :
#   실행문장 2

total = int(input('몇 명이 들어왔나요?'))

if total > 0 :
  print(f'{total}명이 들어왔습니다.')

# 0명이 입력됐을 때 아무것도 뜨지 않음
```

```python
# 0명이 들어왔을 때 잘못된 정보라는 것을 반환하기

total = int(input('몇 명이 들어왔나요?'))

if total > 0 :
  print(f'{total}명이 들어왔습니다.')
else :
  print('잘못된 정보를 입력하였습니다.')
```

```python
# 위에서 했던 날씨 if문에도 예외처리 해주기
wheather = input('오늘 날씨는 어떤가요?')

if wheather != '' :
  print(f'오늘 날씨는 {wheather}입니다.')
else :
  print('날씨가 입력되지 않았습니다.')
```

# 다중 조건문 - if elif else

* 조건이 3개 이상인 경우 사용되고 마무리는 else로 하는 것을 추천

```python
# if 조건1 :
#   실행문장 1
# elif 조건2 :
#   실행문장 2
# elif 조건3 :
#   실행문장 3
# else :
#   실행문장 4

# 조건문이 길어질 때는 조건문에 대한 설명을 주석으로 잘 남기기!
```

```python
# 예제 1
total = int(input('몇 명이 들어왔나요?'))

if total > 5 :
  print(f'{total}명이 들어왔습니다. 많이 들어왔어요!')
elif 5 >= total > 0 :
  print(f'{total}명이 들어왔습니다. 적게 들어왔어요ㅠ')
else :
  print(f'잘못된 정보를 입력하였습니다.')
```

```python
# 예제 2.
wheather = input('오늘 날씨는 어떤가요?')

if wheather == '맑음' :
  print(f'오늘 날씨는 {wheather}입니다. 옷을 얇게 입어도 돼요!')
elif wheather == '비' :
  print(f'오늘 날씨는 {wheather}입니다. 우산을 꼭 챙기세요!')
elif wheather == '눈' :
  print(f'오늘 날씨는 {wheather}입니다. 따뜻하게 입으세요!')
else :
  print('날씨가 입력되지 않았습니다.')
```

# 삼항 연산자

[사용 방법]
* 참일 때의 값 / if / 조건식 / else / 거짓일 때 값
* 변수로 바로 할당할 수 있음

```python
# 삼항 연산자 예시

age = 19

# 참일 때의 값         / if / 조건식 / else / 거짓일 때의 값
is_adult = '성인입니다.' if age > 20 else '미성년자입니다.'

print(is_adult)
```

# 중첩 조건문

* if안에 또 다른 if를 넣어서 다양한 case를 만족시키는 방법

```python
# 조건 1 : Vip 일 때 -> 20번 이상 방문 / 100만원 이상 구매
# 1) 20번 이상 & 100만원 이상
# 2) if 20번 이상
#     if 100만원 이상
# 2번처럼 중첩으로 작성했을 때 직관적으로 눈에 잘 보임
```

```python
# 중첩 조건문 예시

apple = input('사과의 상태를 입력하세요. (좋음 or 나쁨)')

if apple == '좋음' :
  appleprice = int(input('사과 한 개당 가격을 입력하세요.'))
  if appleprice < 1000 :
    print('10개를 산다.')
    print(f'사과 가격은 {appleprice * 10}원 입니다.')
  else :
    print('5개를 산다.')
    print(f'사과 가격은 {appleprice * 5}원 입니다.')
else :
  print('사과를 사지 않습니다.')
```

* B - 80점 이상
* C - 60점 이상
* D - 40졈 이상
* F - 40점 이하 
* 결석이 3번인 경우 무조건 F

중첩 조건문 또는 연산자를 활용하여 if문을 작성해보세요

```python
# 내 답 _ 중첩조건문 사용

grade = int(input('점수를 입력하세요.'))
day = int(input('결석일을 입력하세요.'))

if grade >= 90 :
  if day < 3 :
    print('A입니다.')
  else :
    print('결석 3번으로 F입니다.')
elif 90 > grade >= 80 :
  if day < 3 :
    print('B입니다.')
  else :
    print('결석 3번으로 F입니다.')
elif 80 > grade >= 60 :
  if day < 3 :
    print('C입니다.')
  else :
    print('결석 3번으로 F입니다.')
elif 60 > grade >= 40 :
  if day < 3 :
    print('D입니다.')
  else :
    print('결석 3번으로 F입니다.')
elif grade < 40 :
  if day < 3 :
    print('F입니다.')
  else :
    print('결석 3번으로 F입니다.')
else :
  print('잘못된 정보입니다. 다시 입력해주세요.')
```
