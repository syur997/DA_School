```python
# 리스트 함수를 사용해서 실습해보기

# 상황 1. 버스에 버스기사 아저씨가 탑승해야 운행이 가능

bus = []  # 버스가 비어있음

# 버스기사 아저씨가 탑승을 하려면

bus.append('버스기사 아저씨')  # append로 기사아저씨 태워드리기

print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
```

```python
# 상황 2. 버스에 철수가 탑승을 했다.
bus.append('철수')
print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
```

```python
# 상황 3. 선생님이 아이 3명을 데리고 탑승했어요.
kinder = ['선생님', '아이1', '아이2', '아이3']
bus.extend(kinder)  # list형태로 하나의 리스트에 넣을 때는 extend 활용!

print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
print(f'몇 명이나 타고 있을까요? {len(bus)}명이 타고 있어요.')
```

```python
# 상황 4. 노인 한 분이 탑승하셨어요. 노인은 노인석인 맨 앞자리에 앉으셔야 돼요.
# insert()

bus.insert(1, '노인')

print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
print(f'몇 명이나 타고 있을까요? {len(bus)}명이 타고 있어요.')
```

```python
# 아이의 위치 확인
kid1 = bus.index('아이1')
print(kid1)
```

```python
# 노인분이 내리고자 할 때
bus.remove('노인')

print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
print(f'몇 명이나 타고 있을까요? {len(bus)}명이 타고 있어요.')
```

```python
# 상황 6. 선생님이 아이들을 한 명 한 명 인원 체크를 하며 내리려고 할 때
# 인원체크 후에 선생님이 마지막으로 내릴거에요

bus.pop()  # 맨 마지막을 없앨 때는 pop 사용
print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
print(f'몇 명이나 타고 있을까요? {len(bus)}명이 타고 있어요.')

bus.pop()  # 맨 마지막을 없앨 때는 pop 사용
print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
print(f'몇 명이나 타고 있을까요? {len(bus)}명이 타고 있어요.')

bus.pop()  # 맨 마지막을 없앨 때는 pop 사용
print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
print(f'몇 명이나 타고 있을까요? {len(bus)}명이 타고 있어요.')

bus.pop()  # 맨 마지막을 없앨 때는 pop 사용
print(f'버스에는 누가 타고 있을까요? {bus}가 타고 있어요.')
print(f'몇 명이나 타고 있을까요? {len(bus)}명이 타고 있어요.')
```

# 리스트 컴프리핸션

```python
# 기본 리스트에 값 채워넣기

li = []
for i in range(5) :
  li.append(i)

print(li)
```

```python
# 위 for문을 컴프리핸션 사용해서 한 줄로 줄이기
[i for i in range(5)]
```

```python
# 1) 수식 적용
[i*10 for i in range(5)]  # 0, 1, 2, 3, 4에 10을 곱해줘
```

```python
# 2) 조건문 사용
[i for i in range(5) if i%2 == 0]  # 0~4 중에 나머지가 0인 것만 보여줘
```

```python
# enumerate() 함수
# enumerate된 key값(0, 1, 2)들은 보존이 됨

for i, v in enumerate(['가위','바위', '보']) :
  print(i, v)
```

```python
# enumerate() 함수

for i in enumerate(['가위','바위', '보']) :
  print(i)
```

```python
# zip 함수
# 같은 index자리끼리 결합

alist = ['가위', '바위', '보']
blist = ['가위', '바위', '보']

for a, b in zip(alist, blist) :
  print(a, b)
```

```python
for x, y in zip(range(5), range(5)) :
  print(x, y)
```

```python
# 딕셔너리형태에서도 가능
dict_num = {x:y for x, y in zip(range(5), range(5))}
dict_num
```

```python
# enumerate & zip 함수

alist = ['가위', '바위', '보']
blist = ['가위', '바위', '보']
for i, (a, b) in enumerate((zip(alist, blist))) :
  print(i, a, b)
```

# 3. 딕셔너리

```python
# 딕셔너리 형태
# key : value

# 딕셔너리 기초
x = {}
y = dict()

print(x, type(x))
print(y, type(y))
```

```python
a = {'name' : 'Kim', 'phone' : '01012345678'}
print(a, type(a))
```

```python
# 딕셔너리에 값 추가

a = {'name' : 'Kim', 'phone' : '01012345678'}

a['address'] = 'Seoul'

print(a)
```

```python
# 딕셔너리의 개수
print(a, len(a))
```

## 딕셔너리 함수

```python
# key 값만 가져오는 함수
print(f'key 값은 {a.keys()} 입니다.')

# value 값만 가져오는 함수
print(f'value 값은 {a.values()} 입니다.')

# 전체 다 가져오는 함수
print(f'a 값은 {a.items()} 입니다.')
```

```python
# 딕셔너리 안에 요소 확인하기
'name' in a.keys()
```

```python
'Kim' in a.values()
```

```python
# 값 삭제

del a ['name']  # key 값만 적어주면 value와 함께 삭제됨
print(a)
```

# set(세트)

```python
# 세트 사용
empty_set = set()
print(empty_set, type(empty_set))
```

```python
# 세트에 값 추가
python_set = set(['판다스', '넘파이'])
print(python_set)
```

```python
# 순서를 전혀 고려하지 않는 자료 구조임
test_set = set('python')
print(test_set)
```

```python
# 요소 추가

# 여러 자료 추가
set_1.update([5, 6, 7, 8, 9, 10])
print(set_1)

# 요소 제거_ 특정 순서
set_1.remove(1)
print(set_1)

# 요소 제거_ 특정 값 삭제
set_1.discard(10)
print(set_1)
```

```python
# 세트 연산

s1 = set(['파이썬', 'SAS', 'R', 'SQL'])
s2 = set(['파이썬', '파이토치', '텐서플로', 'SQL'])

print(s1.union(s2))  # union: 합집합

# set은 중복을 허용하지 않기 때문에 중복 제거하고 나오게 됨
```

```python
# 교집합
print(s1.intersection(s2))
```

```python
# 차집합
print(s1.difference(s2))
```
