# set (집합)

* 순서가 없음
* 중복값 없음

```python
# 일반 리스트
foods = ['된장찌개', '피자', '제육볶음'] + ['피자']
foods
```

```python
# foods 리스트를 set형태로 바꿔주기

# 1. 리스트를 변수에 넣어 변수를 set으로 감싸기
foods = ['된장찌개', '피자', '제육볶음']
foods_set1 = set(foods)

# 2. set 안에 리스트 그대로 넣기 
foods_set2 = set(['된장찌개', '피자', '제육볶음'])
```

```python
print(foods_set1)
print(foods_set2)
```

```python
# set에 중복값 넣어보기
foods = ['된장찌개', '피자', '제육볶음', '된장찌개']
foods_set = set(foods)
print(foods)
print(foods_set)
```

```python
# 합집합 -> |
menu1 = set(['된장찌개', '피자', '제육볶음'])
menu2 = set(['된장찌개', '떡국', '김밥'])
menu3 = menu1 | menu2
print(menu3)
```

```python
# 교집합 -> &
menu1 = set(['된장찌개', '피자', '제육볶음'])
menu2 = set(['된장찌개', '떡국', '김밥'])
menu3 = menu1 & menu2
print(menu3)
```

```python
# 차집합 -> -
menu1 = set(['된장찌개', '피자', '제육볶음'])
menu2 = set(['된장찌개', '떡국', '김밥'])
menu3 = menu1 - menu2
print(menu3)
```

# if

* food에 랜덤으로 음식을 하나 넣기
* 만약에 제육볶음이 나오면 곱빼기로 시키기

```python
# food에 랜덤으로 음식 하나 넣기
import random

food = random.choice(['된장찌개', '피자', '제육볶음'])

# 만약에 제육볶음이 나오면 곱빼기로 시키기
if food == '제육볶음' :
  print('곱빼기로 주세요!')
print('주문 완료')
```

* 다른 음식이 나오면 음식 이름 나오게 하기

```python
# food에 랜덤으로 음식 하나 넣기
import random

food = random.choice(['된장찌개', '피자', '제육볶음'])

# 만약에 제육볶음이 나오면 곱빼기로 시키기
if food == '제육볶음' :
  print('곱빼기로 주세요!')
# 다른 음식이 나오면 음식 이름 나오게 하기
else :
  print(food)
print('주문 완료')
```

* q를 누르면 while문이 종료되도록 하기

```python
# while문에서 break 추가하기
lunch = ['된장찌개', '피자', '제육볶음', '짜장면']

while True :
  print(lunch)
  item = input('음식을 추가해주세요 : ')
  if item == 'q' :
    break
  else :
    lunch.append(item)
print(lunch)
```

# 전체 코드

```python
import random
import time

# 메뉴 추가
lunch = ['된장찌개', '피자', '제육볶음', '짜장면']

while True :
  print(lunch)
  item = input('음식을 추가해주세요 : ')
  if item == 'q' :
    break
  else :
    lunch.append(item)
print(lunch)

# 메뉴 삭제
set_lunch = set(lunch)

while True :
  print(set_lunch)
  item = input('빼고 싶은 음식을 말하세요 : ')
  if item == 'q' :
    break
  else :
    set_lunch = set_lunch - set([item])
print(set_lunch)

# 5초 카운트다운
print(set_lunch, '중에서 선택합니다!')
print('5')
time.sleep(1)
print('4')
time.sleep(1)
print('3')
time.sleep(1)
print('2')
time.sleep(1)
print('1')
time.sleep(1)

# 남은 메뉴 중 랜덤으로 음식 추천하기
print(random.choice(list(set_lunch)))
```
