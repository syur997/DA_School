* 질문과 답변을 어떻게 저장할 것인지?

질문 : 취미가 뭐에요?
답변 : 영화보기입니다.

질문 : 좋아하는 음식이 뭐에요?
답변 : 마라탕 좋아해요.

## 1. 온전히 딕셔너리만 쓰기

* 질문은 KEY에
* 대답은 Value에

````python
total_dictionary = {}

while True :
  question = input('질문을 입력해주세요 : ')
  if question == 'q' :
    break
  else :
    total_dictionary[question] = ''

print(total_dictionary)
````

````python
for i in total_dictionary :
  print(i)
  answer = input('답변을 입력해주세요 : ')
  total_dictionary[i] = answer
print(total_dictionary)
````

## 2. 딕셔너리 + 리스트

````python
total_list = []

while True :
  question = input('질문을 입력해주세요 : ')
  if question == 'q' :
    break
  else :
    total_list.append({'질문' : question, '답변' : ''})  # 딕셔너리를 리스트에 추가

print(total_list)

# 질문, 답변이라는 글자가 같이 출력됨
````

````python
for i in total_list :
  print(i['질문'])  # 질문의 value값을 가져와야하기 때문
  answer = input('답변을 입력해주세요 : ')
  i['답변'] = answer
print(total_dictionary)
````
