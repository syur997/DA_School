* 기준 : 범주형 데이터 (성별, 국적, 지역, 혈액형, 학년 등)
* 파악할 데이터: 수치형 데이터



```

```




* 데이터를 그룹별로 분할

2. apply : 분류된 데이터(그룹별)에 함수 적

````python
data = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울', '제주', '제주', '부산', '부산', '제주'],
    '연도': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    '전입신고 수': [10040, 22300, 6000, 12400, 4000, 600, 11100, 45000, 31000],
    '세금': [1004000, 2230000, 600000, 1240000, 400000, 60000, 1110000, 4500000, 3100000]
})

# 여기서 범주형 데이터는 도시, 연도
# 수치형 데이터는 전입신고 수, 세금
````

````python
data
````

````python
# 도시별로 그룹화하여 전입신고 수의 합계 구하기
data.groupby('도시')['전입신고 수'].sum()
````

````python
# 도시별로 그룹화하여 전입신고 수의 평균 구하기
data.groupby('도시')['전입신고 수'].mean()
# [인덱스명]으로 가져와서 시리즈로 결과를 반환함
````

````python
# 도시와 연도별로 그룹화를 하고 싶다면?
# 리스트에 넣기
data.groupby(['도시', '연도']).mean()
# [인덱스명]을 쓰지 않아서 데이터프레임 형태로 반환함
````

````python
# 도시, 연도로 그룹화하여 세금 칼럼의 합계 가져오기
data.groupby(['도시', '연도'])['세금'].sum()
````

# 데이터프레임 병합

## concat

````python
data = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]
    }, index=['a', 'b', 'c', 'd']
)

data2 = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]},
    index=['e', 'f', 'g', 'h']
    )
````

````python
# 두 개의 데이터 합치기
add_data = pd.concat([data, data2])
add_data
````

## ignore_index

````python
# ignore_index = True : 인덱스 재배열 (default는 False로 되어있음)
# ignore_index = True 로 인자값 추가
add_data = pd.concat([data, data2], ignore_index = True)
add_data
````

## axis

````python
data = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]
    }
)

data2 = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]}
    )
````

````python
# axis=1 : 열로 붙음
add_data = pd.concat([data, data2], ignore_index = True, axis=1)
add_data
````

````python
# axis= (default값이) 0 : 행으로 붙음
add_data = pd.concat([data, data2], ignore_index = True, axis=0)
add_data
````

# Join

* inner join : 교집합
* outer join : 합집합
* left join
* right join

````python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                  'F': ['F2', 'F3', 'F6', 'F7']}, index=[2, 3, 6, 7])
````

````python
df4
````

````python
pd.concat([df1, df4], axis=1, join='outer')  # 합집합
````

````python
pd.concat([df1, df4], axis=1, join='inner')  # 교집합
````

````python
pd.concat([df1, df4], axis=1).reindex(df1.index)  # left join
````

````python
pd.concat([df1, df4], axis=1).reindex(df4.index)  # right join
````

## merge
* default : inner join

````python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                  'A': ['A0', 'A4', 'A2', 'A3'],
            'B': ['B0', 'B4', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                  'C': ['C0', 'C1', 'C2', 'C3'],
                  'D': ['D0', 'D1', 'D2', 'D3']})
````

````python
print(left)
print(right)
````

````python
mi = pd.merge(left, right)  # default : inner join
````

````python
print(mi)
````

````python
mo = pd.merge(left, right, how = 'outer')  # outerjoin
print(mo)
````

````python
# join할 때 원하는 값 지정하기
m_key = pd.merge(left, right, on='key')
print(m_key)
````

````python
m_l = pd.merge(left, right, on='key', how='left')
print(m_l)
````

````python
m_r = pd.merge(left, right, on='key', how='right')
print(m_r)
````
