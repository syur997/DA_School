```python
response = requests.get('http://paullab.synology.me/stock.html')
response.encoding = 'utf-8'

# response의 text를 html 변수에 넣기
html = response.text

# hmtl.parser(BeautifulSoup의 Parser 중 하나) -> html문서를 구문 분석, 파싱에 사용
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 그냥 soup을 출력하면 가독성이 떨어짐
print(soup)
```

```python
# prettify -> html 문서형식대로 보기 쉽게 출력됨
print(soup.prettify())
# 그런데 커서를 내리면서 밑에까지 확인하는게 번거로움
```

### ② html을 파일로 저장해서 확인하기

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('http://paullab.synology.me/stock.html')
response.encoding = 'utf-8'
html = response.text

# 파일 쓰기모드로 열어서 가져온 html을 작성하기
f = open('test.hmtl', 'w', encoding= 'utf-8')
f.write(html)
f.close()
```

```python
# 폴더에 어떤 파일 있는지 확인
!dir
# 다운 받은 경로로 들어가면 에디터에서 html을 확인할 수 있음
```

# BeautifulSoup

* str타입의 html 데이터를 html구조를 가진 데이터로 가공해주는 라이브러리
* html을 태그별로 찾을 수 있도록 해줌

  [종류]
* **BeautifulSoup(markup, 'html.parser')**
* BeautifulSoup(markup, 'lxml')
* BeautifulSoup(markup, 'xml')
* BeautifulSoup(markup, 'lxml-xml')
* BeautifulSoup(markup, 'html5lib')

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('http://paullab.synology.me/stock.html')

response.encoding = 'utf-8'
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# title태그 가져오기
soup.title
```

```python
# title태그에서 string만 가져오기
print(soup.title.string)
print(soup.title.text)
```

```python
# soup의 title의 parnet name 가져오기
soup.title.parent.name
```

```python
# soup data의 처음 만나는 table row 가져오기
soup.tr
```

```python
# soup data의 처음 만나는 table data 가져오기
soup.td
```

```python
# 처음 만나는 table head
soup.th
# tr의 결과값 첫 줄과 같은 데이터
```

```python
soup.table
```
