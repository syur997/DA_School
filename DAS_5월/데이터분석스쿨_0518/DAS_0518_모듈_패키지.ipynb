# 모듈

Python 모듈은 파이썬 코드를 담고 있는 파일로, 함수, 클래스, 변수 등의 정의를 포함할 수 있습니다. 모듈은 코드의 재사용과 구성을 용이하게 하기 위해 사용됩니다.

**모듈은 .py 확장자를 가진 파일**로서, 파이썬 스크립트로 작성됩니다.

이 모듈 파일은 다른 파이썬 프로그램에서 **import 문을 사용하여 불러올 수 있습니다.**

이렇게 모듈을 불러옴으로써, 모듈에 포함된 함수, 클래스, 변수 등을 사용할 수 있게 됩니다.

모듈을 사용하면 코드를 논리적인 단위로 분리하여 개발할 수 있습니다.

관련 기능을 가진 코드들을 한 곳에 모아놓고, 필요할 때마다 모듈을 재사용하여 중복을 피하고 코드의 구조를 조직화할 수 있습니다.

모듈은 다른 프로그램에 쉽게 포함시키거나 배포할 수 있으므로 코드의 재사용성과 유지보수성을 높여줍니다.

# 패키지

Python에서 패키지는 관련된 모듈과 리소스를 모아놓은 디렉터리 구조입니다. 

패키지는 모듈을 구성하는 여러 파일들의 논리적인 그룹화를 가능하게 합니다.

패키지는 일반적으로 디렉터리로 표현되며, 디렉터리에는 __ init __.py 파일이 있어야 합니다.

__ init __.py 파일은 해당 디렉터리를 패키지로 인식시키는 역할을 합니다. 패키지에는 모듈 파일과 하위 패키지가 포함될 수 있습니다.

# 모듈이 실행되는지? 안되는지?

* __ name__ 은 현재 모듈의 이름을 값으로 갖는 내장 변수


```
if __name__ == '__main__' :
  pass
else :
  pass
```


* 위의 코드와 같이 작성하게 되면 파일이 실행될 때 if문이 실행되고, __ main __이 아닌 그 모듈의 이름을 가져오게 된다.

pkg_ex 패키지의 seoul.py에서 SeoulPkg() 클래스 정의 아래에 다음과 같이 코드를 쓰면 __ name __변수가 __ main __일 때 모듈을 직접 실행하는 경우에는 if문에서 출력문과 함께 detail() 메서드를 호출하고 그렇지 않으면 else문으로 호출 안내 문구를 넣어봅시다.

* 위처럼 seoul_new.py 추가해주기

````python
# 런타임 초기화 후 실행하기

import pkg_ex.seoul_new

# 외부에서 실행되었다는 else문이 출력됨
````

````python
#파일명 seoul_new.py

class SeoulPkg:
  def detail(self):
    print("[서울 도심 패키지] 호캉스 100만원")

if __name__ == "__main__":
  print("seoul 모듈 직접 실행되었습니다.")
  trip_seoul = SeoulPkg()
  trip_seoul.detail()
else:
  print("외부에서 seoul 모듈 실행되었습니다.")

# seoul_new.py 파일에 있는 class 코드를 복사해서
# 직접 돌리면 if문의 print문이 실행됨
````
