## 함수 정의

```python
# 함수 정의

def add(x, y) :
  value = x + y
  return value

add(3, 4)
# 코랩에서는 결과값이 나오지만 원래는 나오지 않음
```

## 위치 인자, 키워드 인자

```python
# 위치 인자
def greet(name, age):
    print(f"안녕하세요, {name}님! {age}살이시네요.")

greet("철수", 25)
# greet(name, age) 매개변수의 위치와 순서에 맞게 인자가 알아서 들어가는 것
# 위치와 순서를 고려하여 인자값을 넣어야함
```

```python
# 키워드 인자
def greet(name, age):
    print(f"안녕하세요, {name}님! {age}살이시네요.")

greet(age=30, name="영희")
# greet(name, age) 매개변수, 인자의 순서는 상관없이 매개변수=인자 로 직접 지정해주는 것
# 매개변수의 type에 맞춰서 알맞은 인자를 지정해줘야함
```

```python
# 위치 인자 + 키워드 인자
def greet(name, age):
    print(f"안녕하세요, {name}님! {age}살이시네요.")

greet("철수", age=25)
# 인수가 매개변수에 잘 들어갈 수 있도록 위치와 순서를 고려해야함
```

## * args / **kwargs

* 함수 안에서의 * 과 ** 은 패킹 역할

```python
# *args의 예시

def func1(x, y, *args) :
  print(x)
  print(y)
  if args :   # args 값이 True라면(값이 있다면)
    print(args)
```

```python
func1(1, 2)
# x, y까지만 들어가기 때문에 1, 2만 반환됨
```

```python
func1(1, 2, 3, 4, 5)
# 1과 2는 x, y값으로 들어가고 남은 값들은 *args에 의해 패킹하여 튜플로 반환함
```

```python
# **kwargs (keyword argument)

def func2(x, y, **kwargs):
  print(x)
  print(y)
  if kwargs :
    print(kwargs)
```

```python
func2(1, 2)
```

```python
func2(x=1, y=2, z=3, w=4)
# x, y를 제외한 나머지 값들은 딕셔너리 형태로 패킹하여 반환함
```

```python
# *args와 **kwargs의 조합

def func3(a, *args, **kwargs) :
  print(a)
  print(args)
  print(kwargs)
```

```python
func3(1, 100, 200, x = 1, y = 2)
```

* 함수 밖에서의 * 과 ** 은 언패킹 역할

```python
# 리스트, 튜플 언패킹 - *

def func4(x, y, z) :
  print(x, y, z)
  return x + y + z
```

```python
# 주어진 값이 리스트 형태라면?
params = [1, 2, 3]
func4(params)
# 에러 발생함
```

```python
params = [1, 2, 3]
w = func4(*params) # 함수를 호출할 때 *을 붙이면 언패킹이 됨
```

```python
print(w)
# return의 합계도 잘 나옴
```

```python
# 딕셔너리 언패킹 - ** 

def func5(x, y, z) :
  print(x, y, z)
  return x + y + z
```

```python
params = {'x' : 4, 'y' : 5, 'z' : 6}
w = func5(**params)
print(w)
```

```python
# 리스트 + 딕셔너리 언패킹

def func6(x,y,z,a,b):
  print(x,y,z,a,b)
  return x+y+z+a+b

params1 = [1,2,3]
params2 = {"a" : 4, "b" : 5}
w = func6(*params1, **params2)
print(w)
```

## 데코레이터

* 기존 함수를 변형하지 않고 기능을 수정하거나 추가하고 싶을 때 사용

```python
def uppercase_decorator(func):
    def wrapper():
        result = func().upper()
        return result
    return wrapper

@uppercase_decorator
def say_hello():
    return "hello"

print(say_hello())  # 출력: HELLO
```

```python
def my_decorator(func):
    def wrapper():
        print("함수 실행 전")
        func()  # 매개변수 func()을 입력 -> 인수로 받은 함수가 여기서 실행됨
        print("함수 실행 후")
    return wrapper

@my_decorator
def my_function():
    print("안녕하세요!")

my_function()
```

위 예제에서 my_decorator는 데코레이터 함수입니다.

이 함수는 다른 함수를 매개변수로 받고, 해당 함수를 수정하거나 래핑하는 역할을 합니다.

my_function 함수 위에 @my_decorator라는 데코레이터를 표시하여 my_function 함수를 수정합니다.

이렇게 하면 my_function 함수가 my_decorator에 전달되어 wrapper 함수로 래핑되고, 수정된 함수가 반환됩니다.

my_function을 호출하면 my_decorator에 의해 래핑된 함수가 실행됩니다.

wrapper 함수는 함수 실행 전과 후에 추가적인 동작을 수행하고, 최종적으로 my_function을 호출하여 원래의 함수 동작을 실행합니다.

이것은 간단한 데코레이터의 예제이며, 데코레이터를 사용하면 함수의 동작을 수정하거나 추가 기능을 제공할 수 있습니다.








## 제너레이터

* 필요에 따라 값을 반환할 수 있음

파이썬에서는 제너레이터라는 기능도 제공합니다. 

제너레이터를 사용하면 함수를 처리하는 도중에 잠시 중단하고, 

값을 반환할 수 있으며, 이를 yield라고도 부릅니다. 

이후 필요에 따라 처리를 할 수도 있습니다.

제너레이터 함수의 반환값은 generator 타입 변수입니다. 

제너레이터 객체는 next함수를 이용해 다음 값을 반환할 수 있습니다.

```python
def my_generator():
    yield 1
    yield 2
    yield 3

generator = my_generator()

print(next(generator))
print(next(generator))
print(next(generator))
```

위의 예시에서 my_generator는 제너레이터 함수입니다.

제너레이터 함수는 yield 키워드를 사용하여 값을 반환하고 일시적으로 상태를 유지합니다.

my_generator 함수를 호출하면 제너레이터 객체가 생성됩니다.

제너레이터 객체는 이터레이터로서 동작하며, next() 함수를 사용하여 값을 하나씩 가져올 수 있습니다.

next(generator)를 호출하면 제너레이터는 실행되어 yield 키워드를 만날 때까지 실행되며, 값을 반환합니다.

호출할 때마다 다음 yield 문을 만날 때까지 실행이 진행됩니다.

위 예시에서는 제너레이터를 호출한 후 next() 함수를 세 번 호출하여 각각 1, 2, 3을 순서대로 가져왔습니다.

제너레이터는 값을 생성하는 데 유용하며, 대량의 데이터를 한 번에 메모리에 로드하지 않고도 순차적으로 처리할 수 있습니다.

```python
# yield 가 없을 때는 print문이 한 번에 실행됨
def sample_generator():
  """제너레이터 함수"""
  print("처리 시작")
  print("처리 재개")
  print("처리 재개")

gen_obj = sample_generator() #제너레이터 객체 생성
```

```python
# yield를 print문 중간마다 하나씩 넣고 next를 이용하여 print하면
# yield에서 끊어서 출력함
def sample_generator():
  """제너레이터 함수"""
  print("처리 시작")
  yield "아침 먹자"
  print("처리 재개")
  yield "점심 먹자"
  print("처리 재개")
  yield "저녁 먹자"

gen_obj = sample_generator() # 제너레이터 객체 생성
print(next(gen_obj))
# 첫 번째 yield 까지 출력
```

```python
# 두 번째 yield 실행
print(next(gen_obj))
```

```python
# 세 번째 yield 실행
print(next(gen_obj))
```
