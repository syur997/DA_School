```python
# 함수 예제 14 - 세 개의 숫자를 입력받아 가장 큰 수를 출력하는 print_max 함수 만들기
# 단, if문을 사용해서 수를 비교하세요.

# 내가 쓴거
def print_max(a, b, c):
  if a > b and a > c :
    print(a)
  elif b > c and b > a :
    print(b)
  else :
    print(c)

print_max(1, 5, 4)
```

```python
# 강사님이 쓴거
def print_max(a, b, c) :
  max_val = 0
  if a > max_val :  # 0보다 큰 값을 받기 원함
    max_val = a

  if b > max_val :  # a값이 들어간 max_val이 b보다 작다면 b로 대체됨
    max_val = b
  
  if c > max_val :  # b값이 들어간 max_val이 c보다 작다면 c로 대체됨
    max_val = c

  print(max_val)

print_max(10, 50, 100)
```

```python
# 함수 예제 15 - 입력된 문자열을 역순으로 출력하는 reverse 함수

def print_reverse(string_a):
  print(string_a[::-1])

print_reverse('python')
```

```python
# 함수 예제 16 - 평균을 출력하는 함수

def print_score(score_list):
  print(sum(score_list) / len(score_list))

print_score([1, 2, 3])
```

```python
# 함수 예제 17 - 하나의 리스트를 입력받아서 짝수만 출력하는 함수 만들기

# 내가 쓴거

list_l = []

def print_even(even_list) :
  for i in even_list :
    if i % 2 == 0 :
      list_l.append(i)
    else :
      pass
  print(list_l)

print_even([1, 2, 3, 4, 5, 6, 7, 8, 10])
```

```python
# 강사님이 쓴거

def print_even(print_list) :
  for v in print_list :
    if v % 2 == 0 :
      print(v)

print_even([1, 2, 3, 4, 5])
```

```python
# 함수 예제 19 - 함수 내부의 결과는 접근이 불가하다.

def n_plus(n) :
  result = n + 1

print(n_plus(3))

# 여기서 None이 나오는 이유는 return이 없어서!
# return이 없으면 디폴트는 None값!
```

```python
# 함수 예제 20 - 문자열 하나를 입력 받아서 인터넷 주소를 반환하는 함수 만들기

def make_url(string) :
  url = 'www.' + string + '.com'
  return url

make_url('naver')
```

```python
# 함수 예제 21

def make_list(string) :
  my_list = []
  for i in string :
    my_list.append(i)
  return my_list

make_list('abcd')
```

```python
# 함수 예제 22 - 문자열 대체

def convert_int(string) :
  return int(string.replace(',', ''))  # replace는 ','를 ''으로 바꿈

convert_int('123,456')
```

```python
# 함수 예제 28 - 함수의 호출 과정(키워드 인수)
# 위치인수는 호출할 때 함수의 정의와 인수의 순서가 같은 순서로 지정되어야함
# 위치가 틀어지면 값이 완전히 바뀔 수 있다.

def mul(a, b):
  c = a * b
  return c

mul(3, 5)

def billing(unit_prices, units, fees, tax_rate) :
  return(unit_prices * units + fees) * tax_rate

x = billing(1000, 10, 500, 1.1)
print(x)
```

```python
# 함수 예제 29 - 키워드 함수

def billing(unit_prices, units, fees, tax_rate) :
  return(unit_prices * units + fees) * tax_rate

x = billing(unit_prices = 1000, units = 10, fees = 500, tax_rate = 1.1)

print(x)
```

```python
# 키워드로 지정해주어서 파라미터 위치를 바꿔도 괜찮음
y = billing(units = 10, fees = 500, unit_prices = 1000, tax_rate = 1.1)
print(y)
```

```python
# 위치인수 + 키워드 혼합

x1 = billing(1000, 10, tax_rate = 1.1, fees = 500)
print(x1)
```

```python
# 위치인수 + 키워드를 혼합할 때 : 키워드 + 위치인수는 안됨
y1 = billing(1000, 10, tax_rate = 1.1, 500)
print(y1)
```
