# Join - 문자열 연결

# replace - 문자열 치환하기

* .replace([제거할 것], [변경할 것])
* 문자열에는 replace라는 치환 메서드가 있다.

```python
# 공백을 없애는 replace
text1 = 'Simple is Best'
text2 = text1. replace(' ', '')
print(text1)
print(text2)
```

```python
# replace에 숫자 넣기
text1 = 'Simple is Best Plan'
text2 = text1. replace(' ', '__', 2)  # 넣은 수만큼 앞에서부터 바꿔줌
print(text1)
print(text2)
```

# 문자열 공백 처리

* strip함수를 사용하여 공백을 제거할 수 있다.
* 앞 뒤 공백을 제거할 수 있다.

```python
# 공백을 제거하지 않았을 때
text = ' abcdefg '
print('*' + stripped + '*')
```

```python
# strip을 사용하여 공백을 제거했을 때
text = ' abcdefg '
stripped = text.strip()
print('*' + stripped + '*')
```

```python
text = ' abcdefg '

# 오른쪽 공백 제거 - rstrip
r_stripped = text.rstrip()
print(r_stripped)

# 왼쪽 공백 제거 - lstrip
l_stripped = text.lstrip()
print(l_stripped)
```

* zfill 메서드를 사용하면 인수로 지정한 길이가 되도록 0으로 채운다.

```python
# CSV 데이터를 읽을 때,
# 0000092 -> 문자열인데 파이썬에서 읽으면 92라는 숫자로 인식해서 92로 읽어옴
# 원래의 데이터인 0000092로 바꿔주기 위해 zfill을 사용하여 0으로 채워줄 수 있음

num = 92   # 숫자열을
num_str = '92'  # 문자열로 바꾼 후
zfilled = num_str.zfill(6)  # 0으로 채워주기
print(zfilled, type(zfilled))
```

# 특정 문자열 포함 행만 추출하기

* 문자열의 split과 리스트 컴프리핸션을 사용하면 특정 문자열을 포함하는 행만 추출 가능

1. split을 사용하여 텍스트를 줄바꿈, 1행씩 리스트로 얻을 수 있음
2. 리스트 컴프리핸션을 사용하여 특정 문자열을 포함하는 행 추출
3. 원래 텍스트로부터 특정한 문자열을 포함하는 행만 추출한 텍스트를 만들 수 있다.

```python
# 미션 - '데이터' 글자가 있는 행만 가져오고 싶다.

text = """저는 김철수입니다.
        반갑습니다. 저는 데이터 분석가입니다.
        데이터로 일하는 것은 즐겁습니다.
        더 열심히 배우겠습니다. 감사합니다. """
```

```python
# 한줄씩 리스트로 생성
lines = text.split("\n")  # 줄바꿈 기준으로 split
lines
```

```python
# 라인에서 '데이터'가 있는 라인만 가져오기
line_list = [line for line in lines if "데이터" in line]  # 한 줄씩 가져와서 for문으로 돌리며 '데이터'가 있는 줄만 가져오기
line_list
```

```python
# 문자열 리스트로 연결하기
# enter로 연결하여 출력
new_text = "\n".join(line_list)
print(new_text)

# 결과물에서 공백을 제거하고 싶다면 lstrip 활용
```

# 텍스트의 빈 행 삭제하기

```python
text = """저는 김철수입니다.

        반갑습니다. 저는 데이터 분석가입니다.

        데이터로 일하는 것은 즐겁습니다.
        
        더 열심히 배우겠습니다. 감사합니다. """
```

```python
# 줄바꿈 기준으로 split하면 공백행도 같이 구분되어 추출됨
lines = text.split('\n')
lines
```

```python
line_list = [line for line in lines if line.strip() != '']
print(line_list)
```

```python
new_text = '\n'.join(line_list)
print(new_text)
```

```python
# 위 결과물이 깔끔하지 못해서 깔끔하게 공백을 제거하고 싶다면?

line_list2 = [line.strip() for line in lines if line.strip() != '']  # line에 .strip() 해주면 됨
print(line_list2)

new_text2 = '\n'.join(line_list2)
print(new_text2)

# 결과물이 훨씬 깔-끔 !
```

```python
# print(new_text.strip())으로 했을 때는
# new_text가 각 문장별로 구분되지 않고 join으로 묶여서 하나의 문장이 되었기 때문에
# 모든 공백이 지워지는게 아니라 new_text결과물에서 맨 앞과 맨 뒤의 공백만 지워짐

print(new_text.strip())
```
