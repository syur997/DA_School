## 1. Series (시리즈)

- 인덱스를 지정할 수 있음

[ 시리즈를 만들 때 사용할 수 있는 방법 ]
1. 리스트
2. 딕셔너리
3. 스칼라

````python
import pandas as pd
````

### 리스트로 시리즈 만들기
* 인덱스명을 지정하지 않으면 자동으로 생성됨

````python
s = pd.Series([1, 2, 3, 4])
print(s)
# 시리즈를 만들면 첫 열은 자동으로 index가 생성됨
````

### 딕셔너리로 시리즈 만들기
* index명을 지정할 수 있음

````python
s2 = pd.Series({'a': 1, 'b': 2, 'c': 3, 'd': 4})  # index명은 알파벳도 가능
print(s2)
````

### 스칼라값으로 시리즈 만들기
* 5라는 스칼라값 하나로 index지정해서 시리즈 만들기

````python
s3 = pd.Series(5, index = [0, 1, 2, 3, 4])
print(s3)
````

### 변수로 시리즈 만들기

* 데이터를 변수에 넣어서 시리즈를 만들 수 있음
```
pd.Series(변수, index=변수)
```
  --> 유지보수가 쉽고 한 눈에 보기 편함


````python
data = [1, 2, 3, 4, 5]
index = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=index)
print(series)
````

* 인덱스를 지정하지 않아도 자동 생성

````python
s = pd.Series(data)
print(s)
````

## 인덱싱, 슬라이싱

````python
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series)
````

````python
series[0]
````

````python
series[0:4]
````

````python
series[::-1]
````

* 슬라이싱을 아무리해도 원본은 바뀌지 않음

## Boolean Indexing

````python
print(series)
````

* print문에서 시리즈에 조건을 추가했을 때

````python
print(series > 2)
````

* 위처럼 True or False가 아니라 시리즈로 조건에 맞는 값을 받고 싶을 때


```
시리즈명[시리즈명 조건]
```



````python
print(series[series > 2])  # a[a조건]
````

````python
# 5보다 작은 시리즈 만들기
print(series[series < 5])
````

## Series 산술연산

````python
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
````

````python
print('더하기 결과')
print(series + 100)

print('------------------------')
print('빼기 결과')
print(series - 100)

print('------------------------')
print('곱하기 결과')
print(series * 100)

print('------------------------')
print('제곱 결과')
print(series ** 100)

print('------------------------')
print('나누기 결과')
print(series / 100)

print('------------------------')
print('나머지 결과')
print(series % 100)

print('------------------------')
print('몫 결과')
print(series // 100)
````

### loc - 행에 접근
* loc은 숫자가 들어갈 수 없고 label만 들어갈 수 있음
* 만약 index가 숫자로 되어있으면 숫자로 쓰면 됨

````python
s = pd.Series([10, 20, 30, 40], index = ['a', 'b', 'c', 'd'])
print(s)
````

````python
print(s.loc['b'])
````

* 숫자가 아닌 인덱스명으로 슬라이싱을 할 때는 종료 인덱스를 포함하여 출력함

````python
print(s.loc['b':'d'])
````

### iloc - 열에 접근
* **행에 기반한** 열에 접근

````python
# 1 -> 두번째 행의 열 값에 접근
print(s.iloc[1])
````

````python
print(s.iloc[1:3])
````

````python
print(s.iloc[-1])
````

### at
* 특정 인덱스를 지정하여 값을 가져올 때
* loc이랑 비슷하지만 하나의 값만 가져올 때 사용

````python
print(s)
````

````python
print(s.at['a'])
````

### iat
* iloc처럼 행에 기반하여 열에 접근

````python
print(s.iat[0])
````

### get
* 없는 값에 대한 처리 방법

````python
# loc에서 없는 값을 부를 때
print(s.loc['e'])
# 에러 발생
````

````python
# at에서 없는 값을 부를 때
print(s.at['e'])
# 에러 발생
````

````python
# get에서 없는 값을 부를 때
print(s.get('e'))
# 에러 없이 None 출력
````

````python
# default값을 지정해주면 값이 없어도 에러가 나지 않고 대체값을 출력해줌
print(s.get('e', default=1))
````

## 결측값 (NaN, None)
- NaN (Not a Number) : 데이터 타입이 float
- None : 데이터 타입이 None

[ 확인하는 법 ]
- isna()
- isnull()

````python
import numpy as np
````

* 파이썬에서의 None이 Numpy와 Pandas에서 각각 어떻게 적용될까?

````python
# 파이썬에서 None값을 가진 리스트를 생성
data = [1, 2, 3, None]
````

````python
# Numpy에서는 그대로 None으로 반환
print(np.array(data))
````

````python
# Pandas는 NaN(float)으로 변형하여 출력
print(pd.Series(data))
````

````python
s = pd.Series(data)
````

* isna() 와 isnull() 사용해서 확인해보기
* nan, null값일 때 -> True

````python
print(s.isna())
print(s.isnull())
````

````python
# null값 갯수 세기
print(s.isna().sum())
print(s.isnull().sum())
````

````python
print(s.notnull())
````

## 결측치 제거

* 결측치 제거

````python
s.dropna()
````

* 결측치 치환

````python
s
````

````python
s.fillna(0)
# 숫자, 문자열 다 가능
````

## Series 함수

````python
s = pd.Series([1, 3, 5, 7, 9])
````

````python
s.count()  # 카운트함수
````

````python
s.min()  # 최솟값
````

````python
s.max()  # 최댓값
````

````python
print(s.mean())  # 평균
print(s.median())  # 중앙값
print(s.std())  # 표준편차
print(s.var())  # 분산
````
