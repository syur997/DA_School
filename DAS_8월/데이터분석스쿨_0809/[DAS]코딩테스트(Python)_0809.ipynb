# 문제2.
- < pass > 부분에 리스트 내장함수를 insert를 이용하여 코드를 입력하고 다음과 같이 출력되게 하세요.

```
l = [200, 100, 300]
<pass>

```

````python
# 정답 1
l = [200, 100, 300]
l.insert(2, 10000)
print(l)
````

````python
# 정답 2
l = [200, 100, 300]
l[:2] + [10000]
````
