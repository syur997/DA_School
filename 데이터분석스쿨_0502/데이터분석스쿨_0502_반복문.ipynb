```python
# 반복문 기초 break - continue

# 반복 출력

for i in range(10) :
  if i == 3 :
    break
  if i == 1 :
    continue
print(i)

# 첫 번째 if문의 조건인 i == 3 에 만족해서 i는 3인채로 코드가 멈추게 됨
# 그래서 마지막 print(i)에는 3이 출력됨
```

```python
# 위 코드 응용

for i in range(5) :  # [0, 1, 2, 3, 4]

  # 0, 2는 if문 조건에 맞지 않아서
  if i == 3 :
    break
  # 1은 두번째 if문에서 걸려서 print로 가지 않고 continue로 감
  if i == 1 :
    continue

  # 걸리지 않고 바로 print문으로 내려옴
  print(i)  # 0, 2
```

# break

* break는 흐름을 끊어 중단할 때 사용한다.
* 여기서 주의하셔야 할 점은 이 break 구문은 바로 위의 for나 while문만 탈출한다는 것!

* 반복문(loop) 순회 중에 break 문을 만나면 반복문의 내부 블록을 벗어나게 된다. 그러나 아래의 예제 1번과 같이 어떠한 장치 없이 break를 넣으면 반복문이 반복하지 않으므로 예제 2번과 같이 주로 조건 안에 넣어 실행한다.

```python
# 예제1(while)
while 조건:
	반복 실행할 코드
	break # while 구문을 탈출

# 예제1(for)
for 변수 in 범위:
	break # for 반복문 탈출
```

```python
# 예제2(while)
while 조건:
	반복 실행할 코드
	if 조건:
		break # while 구문을 탈출

# 예제2(for)
for 변수 in 범위:
	반복 실행할 코드
	if 조건 :
		break # for 반복문 탈출	
```

## while문 문제3

```python
# 1~100 사이의 숫자를 맞출 때까지 반복하는 while문을 작성
# 내가 입력한 숫자가 컴퓨터숫자보다 작으면 '값이 작습니다.'
# 내가 입력한 숫자가 컴퓨터숫자보다 크면 '값이 큽니다.'
# 답이 맞으면 '맞습니다.'

import random
guess_number = random.randint(1,100)
user_input = int(input("숫자를 입력해서 값을 맞춰보세요"))

while (user_input != guess_number):
  if user_input > guess_number:
    print("숫자가 큽니다.")
  else:
    print("숫자가 작습니다.")
  user_input = int(input("숫자를 입력해서 값을 다시 맞춰보세요"))
else:
  print("정답입니다.")
```

2. 손님별 운행시간은 5 ~ 20분으로 정합니다.
3. 그 중 5 ~ 15분인 고객의 콜을 받습니다.
4. 기본요금은 기본 5분일 경우 5,000원으로 1분이 추가될때마다 1,000원을 더 추가합니다.
5. 1) 오늘 콜을 받은 횟수와 2) 탑승횟수, 3) 30만원을 채우기까지 총 소요시간 그리고 4) 승객에게 받은 요금과 5) 사납금을 납부하고 남은 금액을 최종 산출하세요.

```python
# 내가 풀어보던거...

import random

time = random.randint(5, 20)
price = 0

while 
if 15 >= time >= 5 :
  price = (time // 5 * 5000) + (time % 5 * 1000)
  print(f'운행시간은 {time}분이고 요금은 {price}원입니다.')
else :
  print('운행시간을 초과했습니다.')
```

```python
# 강사님 답

import random

# 매칭횟수(콜 받은 횟수)
matching = 0
# 실제로 탄 고객(탑승횟수)
people = 0
# 택시 이용요금(승객에게 받은 요금)
money = 0
# 총 운행 시간(총 소요시간)
all_time = 0

# 운행은 30만원 이상일 때 종료
while money < 300000 :
  matching += 1
  time = random.randint(5, 20)

  if 5 <= time <= 15 :
    people += 1
    each_money = time * 1000
    print(f'손님 {people}번이 탑승했습니다. 소요시간은 {time}분이 걸리고 택시요금은 {each_money}원입니다.')
    money += each_money
    all_time += time
  else :
    print('매칭에 실패했습니다.')

print(f'오늘 매칭 회수는 {matching}번이고, 탑승 손님은 {people}명이고, 소요시간은 {all_time}분이고, 총 매출은 {money}원입니다.')
```
