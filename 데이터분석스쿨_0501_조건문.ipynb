# 1.2 조건문 복습

조건문은 다음과 같이 복습하려고 합니다.

1. 단일 조건문
2. If - else 조건문
3. If - elif - else 조건문
4. 다중 중첩 조건문
5. input과 함께 사용하는 조건문

[기억해야할것]

1. if 뒤에는 참과 거짓을 판단하는 조건이 와야함
2. 조건문이 끝나면 반드시 콜론으로 끝나야함
3. if 다음 문장은 들여쓰기를 사용해야함
4. if 조건에 부합하지 않으면 elif나 else를 사용한다. 보통 조건에 모두 부합하지 않을때 else를 사용한다.

* x == y 같다

* x >= y 크거나 같다
* x <= y 작거나 같다

```python
# if문 예제

x = 2
if x > 1 and x < 10:  # 2가지 조건이 모두 충족 됐을 때
	print('True')
```

```python
# 예측해보기

if True:
    print('hello') 
print('one')

if False:
    print('world')
print('two')
```

```python
# 해설

if True:
    print('hello')  # if문의 조건이 True이기 때문에 'hello'를 출력함
print('one')  # 위의 if문과는 상관 없음, 그냥 print문임

if False:
    print('world')  # if문의 조건이 False이기 때문에 'world'를 출력하지 않음
print('two')  # 위의 if문과는 상관 없음, 그냥 print문임

# 즉, 코드는 [if문 - print문 - if문 - print문]으로 4개의 코드가 있는 것임
# 그래서 결과는 hello, one, two가 결과값으로 나옴
```

```python
# if문에서 0,1로 T,F 표현 가능

if 1:
    print('hello')
print('end')

if 0:
    print('hello')
print('end')

if -1:  # -1은 True
    print('hello')
print('end')
```

```python
#공백이 있는 문자열 -> True
if  ' ':
    print('hello')
print('end')

#공백이 없는 문자열 -> False
if '':
    print('hello')
print('end')
```

```python
if not '':  # ''이 False이기 때문에 not '' 은 True
    print('hello')
print('end')

if None: # None은 False와 같은 의미라고 생각하시면 됩니다.
    print('hello')
print('end')

if not None:
    print('hello')
print('end')
```

# if, else
* if가 거짓일 때, else로 넘어감

```python
x = int(input('비교할 수를 입력하세요 :'))
y = int(input('비교할 수를 입력하세요 :'))

if x > y:
    print('첫번째 입력한 값이 더 큽니다.')
else:
    print('두번째 입력한 값이 더 크거나 같습니다.')
```

```python
x = int(input('비교할 수를 입력하세요 :'))
y = int(input('비교할 수를 입력하세요 :'))

if x > y :
    print('첫번째 입력한 값이 더 큽니다.')
elif x < y :
    print('두번째 입력한 값이 더 큽니다.')
else :
    print('두 수가 같습니다.')
```

```python
# if, elif, else 예제

score = 85
money = 0

if score >= 90:
    print('mom : Excellent!')
    money += 1000000
elif score >= 80:
    print('mom : Great!')
    money += 100000
elif score >= 70 and score < 80:
    print('mom : Good Job!')
    money += 10000
elif score >= 60 and score < 70:
    print('mom : gooood!')
    money += 1000
else:
    print('mom : Oops!')

print(f'{money}원')
```

```python
# 위 코드에서 elif문을 if문으로 바꾸면 
score = -100

if score >= 90:
    print('mom : i\'m so happy!')
    money += 1000000
else:
    print('mom : i\'m...!')
```

```python
# 중첩조건문
# if를 중첩해서 사용하기

score = 98

if score >= 90:
    print('mom : i\'m so happy!')
    money += 1000000
    if score > 95:
        print('hello world')
        if score > 97:
            print('hello!!!')
elif score >= 80:
    print('mom : i\'m happy!')
    money += 100000
```
