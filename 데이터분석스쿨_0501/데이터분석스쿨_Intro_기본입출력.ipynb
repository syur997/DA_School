# 1. 주석

주석(Comment) - 가장 중요하면서도 가장 많이 잊어버리는 것.
만약에 코드가 100줄 이상되는 것이 주어졌는데 아무런 설명이 없다면 개인적으로도 당황스럽고, 업무적으로도 다시 그 코드를 분석하기 위해 시간이 낭비될 수 있음

 주석은 보통 코드가 어떤 내용을 포함하고 있는지, 이 문장이 의미하는 것이 무엇인지, 왜 이렇게 썼는지, 주의할 점은 없는지, 추가로 알아야 할 내용이 있는지에 대한 역할을 하게 된다.
 
 주석은 프로그램을 사용하는데 아무런 영향을 끼치지는 않지만 많은 줄의 코드보다 더 도움이 되는건 한 줄의 주석이 될 수 있기 때문에 앞으로도 주석은 잘 사용해야함

```python
# 실습2. 보드게임 주사위 2개 던지기 -> 난수 추첨

random.seed(16)
die1 = random.sample(range(1, 7), 1)
die2 = random.sample(range(1, 7), 1)

print(die1, die2)

# die1 과 die2의 합을 출력하는 프로그램을 만드세요
total = sum(die1+die2)
total
```

```python
# 위의 코드를 하나의 함수로 변환

def die():
  random.seed(16)
  die1 = random.randint(1, 7)
  die2 = random.randint(1, 7)

  return die1 + die2

die()
```

# 파이썬 멈추기

```python
import time # sleep 함수 사용
```

```python
# 1. 초 단위 테스트
print('-' * 30)
now = datetime.now(timezone(timedelta(hours=9)))
print(now.strftime('%H:%M:%S')) # 현재 시간 출력

time.sleep(5)  # 5초 쉬기

print('-' * 30)
now = datetime.now(timezone(timedelta(hours=9)))
print(now.strftime('%H:%M:%S')) # 현재 시간 출력

time.sleep(5)  # 10초 쉬기
```
