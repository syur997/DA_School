## 예제 1

```python
# class Dog(object): 
class Dog:
  pass
```

```python
# 예제 1 - 틀을 만들자

class Dog:
  species = 'firstdog'   # 고정된 값 (input값에 따라 변하지 않음)

  # 초기화/인스턴스 속성

  # 메서드 = 함수
  def __init__(self, name, age):   # 들어오는 값에 따라 결과값이 바뀜
    self.name = name
    self.age = age

# 클래스 정보
print(Dog) 

# 인스턴스화 - 설계화가 된 것 생성
a = Dog("mikky", 2)
b = Dog("baby", 3)
```

```python
# 클래스 변수 - 직접 접근 가능
# 인스턴스 변수 - 객체마다 별도 존재

# 인스턴스 속성 확인
print(f'{a.name} is {a.age} and {b.name} is {b.age}')
```

```python
# 클래스 변수 - 직접 접근 가능
if a.species == "firstdog" :
  print(f'{a.name} is a {a.species}')

if b.species == "firstdog" :
  print(f'{b.name} is a {b.species}')
```

```python
# 이런 형태로도 접근 가능하다.
# 클래스로도 접근 / 인스턴스화로도 접근 가능

print(Dog.species)
print(a.species)
print(b.species)

# 사실 이 부분이 이해가 되면 클래스가 이해가 된 것

# 나만의 공간, 나만의 속성, 공유하는 공간, 공유하는 속성에 대해 이해가 되었기를 !
```

## 예제 2

```python
# 클래스 변수, 인스턴스 변수

# Warehouse 클래스
class Warehouse:
  # 클래스 변수
  stock_num = 0 # 재고

  def __init__(self, name):
    #인스턴스 변수 # 추가
    self.name = name
    Warehouse.stock_num += 1

  def __del__(self): # 소멸
    Warehouse.stock_num -= 1

user1 = Warehouse('Lee')
user2 = Warehouse('Kim')

# 코드를 실행하면
# __init__의 name에 'Lee'값이 들어가면서 __init__이 자동실행됨
```

```python
print(Warehouse.stock_num)
```

```python
# 다양하게 호출해보기
print(user1.name)
print(user2.name)

# Warehouse class의 정보 보기
print(Warehouse.__dict__)
```

```python
# __del__
del user1

# del로 user1을 삭제하여서 user1에 대한 정보가 사라지고 stock_num도 1로 바뀜
print(Warehouse.__dict__)
print(Warehouse.stock_num)
```

## 예제 3

```python
#예제 3 - return의 사용과 self가 아닌 변수

class Dog:
  species = 'firstdog'

  #초기화/인스턴스 속성

  def __init__(self, name, age):
    self.name = name
    self.age = age

  def info(self):
    return f'{self.name}은 {self.age}살 이에요'

  def speak(self, sound):
    return f"{self.name}은 {sound} 소리를 내요"

#인스턴스 생성
c = Dog('July', 4)
d = Dog('Dorri', 10)

#메서드 호출
print(c.info())
print(d.info())

#메서드 호출
print(c.speak('왈왈'))
print(d.speak('멍멍'))
```

```python
# info에 self가 있기에 값을 가져올 수 있음
print(c.info())
print(d.info())
```

```python
# speak에도 self가 있어서 값을 가져올 수 있음
print(c.speak('왈왈'))
```

```python
# 예제 3
class FishCakeMaker:
  def __init__(self, **kwargs):
    
    self.size = 10
    self.flavor = "팥"
    self.price = 1000

    if "size" in kwargs:
      self.size = kwargs.get("size")
    if "flavor" in kwargs:
      self.flavor = kwargs.get("flavor")
    if "price" in kwargs:
      self.price = kwargs.get("price")

  def show(self):
    
    print(f"붕어빵 종류 {self.flavor}")
    print(f"붕어빵 크기 {self.size}")
    print(f"붕어빵 가격 {self.price}")

fish1 = FishCakeMaker()
fish2 = FishCakeMaker(size = 20, price = 2000)
fish3 = FishCakeMaker(size = 20, price = 3000, flavor = "피자")

fish1.show()
fish2.show()
fish3.show()
```

## 함수와 클래스의 차이점

```python
def add(num1, num2):
    return num1 + num2

def subtract(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    if num2 != 0:
        return num1 / num2
    else:
        print("0으로 나눌 수 없습니다.")

# 사용자로부터 두 개의 숫자 입력 받기
number1 = float(input("첫 번째 숫자를 입력하세요: "))
number2 = float(input("두 번째 숫자를 입력하세요: "))

# 덧셈
result = add(number1, number2)
print("덧셈 결과:", result)

# 뺄셈
result = subtract(number1, number2)
print("뺄셈 결과:", result)

# 곱셈
result = multiply(number1, number2)
print("곱셈 결과:", result)

# 나눗셈
divide(number1, number2)
```

```python
# 위의 함수로 구현한 계산기 코드를 클래스로 바꿔보기

class Operation:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def multiply(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 != 0:
            return self.num1 / self.num2
        else:
            print("0으로 나눌 수 없습니다.")

# 사용자로부터 두 개의 숫자 입력 받기
number1 = float(input("첫 번째 숫자를 입력하세요: "))
number2 = float(input("두 번째 숫자를 입력하세요: "))

# Operation 클래스의 인스턴스 생성
operation = Operation(number1, number2)

# 덧셈 결과 출력
add_result = operation.add()
print("덧셈 결과:", add_result)

# 뺄셈 결과 출력
subtract_result = operation.subtract()
print("뺄셈 결과:", subtract_result)

# 곱셈 결과 출력
multiply_result = operation.multiply()
print("곱셈 결과:", multiply_result)

# 나눗셈 결과 출력
operation.divide()
```
